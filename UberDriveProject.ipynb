## Q19. Plot a bar graph of Purpose vs Miles(Distance). What can you infer from the plot(2 +2 points)
#### Note: Use the original dataframe without dropping the 'NA' values.
#### Hint:You have to plot total/sum miles per purpose

```python
k3 = UberDrives_Data.groupby('PURPOSE*')['MILES*'].sum().sort_values(ascending=False).head(10) 
k3
k3= k3.reset_index() # flatten the dataframe 
k3
k3.columns = ['PURPOSE*' ,'sum_of_miles']
k3
%matplotlib inline 
import seaborn as sns
sns.barplot(data= k3 , x= 'PURPOSE*' , y ='sum_of_miles')
plt.xticks(rotation=90)
```

### The plot shows the total miles driven for different purposes. 
### The most common purpose is meeting followed by customer visit and then meal/entertain. 
### The least common purpose is charity followed by airport/travel, moving and commute. 
### The plot suggests that people drive the most miles for work-related purposes, followed by personal errands and then social events.
